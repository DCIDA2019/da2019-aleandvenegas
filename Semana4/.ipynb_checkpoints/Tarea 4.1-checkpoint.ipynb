```python
import numpy as np
import pandas as pd
import requests
import matplotlib.pyplot as plt
import os
import math
import scipy
import seaborn as sns
import scipy.stats
```

# Dada la distribucion gauciana obtendremos su cdf y lo coprobaremos con histograma #

```python
xmin=66
xmax=74
mean=70
std=2

x=np.linspace(xmin,xmax,10000)
yp=scipy.stats.norm.pdf(x,mean,std)
plt.plot(x,yp)
```

```python
yc=scipy.stats.norm.cdf(x,mean,std)
plt.plot(x,yc,markersize=1)
```

```python
r=np.random.uniform(.05,.95,10000)
#r=np.linspace(.05,.95, 10000)
r.sort()
```

```python
interl=scipy.interpolate.interp1d(yc, x)
xi=interl(r)
```

```python
#plt.plot(xi,yp)
plt.hist(xi, bins=9, density =True)
```

```python
sns.set(palette='nipy_spectral')
sns.distplot(xi, bins = 13)
```

## Algo con lo que me atore un rato fue q con 100 puntos random no se aprecia la gausiana y pensaba q tenia un error pero es simplemente q necesita muchos datos para poder observar como el porcentaje de que caiga en la media es mayor q es los otro puntos ##

```python
x=np.linspace(xmin,xmax,10000)
yc=scipy.stats.norm.cdf(x,mean,std)
r=np.random.uniform(.05,.95,100)
interl=scipy.interpolate.interp1d(yc, x)
xi=interl(r)
import seaborn as sns
sns.set(palette='nipy_spectral')
sns.distplot(xi, bins = 13)
```

# La de poisson #

```python
xmin=0
xmax=30
mu=5

x=np.linspace(xmin,xmax-1,xmax)
yp=scipy.stats.poisson.pmf(x,mu)
plt.plot(x,yp)
```

```python
yc=scipy.stats.poisson.cdf(x,mu)
plt.plot(x,yc,'o')
```

```python
plt.plot(yc,x,'o')
```

```python
r=np.random.uniform(.01,1,10000)
r.sort()
interl=scipy.interpolate.interp1d(yc, x)
xi=interl(r)
```

```python
plt.hist(xi,bins=30)
```

```python
sns.distplot(xi, bins = 17)
```

## Tambien aqui sin muchos datos no se nota la distribucion

```python
x=np.linspace(xmin,xmax-1,xmax)
yc=scipy.stats.poisson.cdf(x,mu)
r=np.random.uniform(.01,1,100)
r.sort()
interl=scipy.interpolate.interp1d(yc, x)
xi=interl(r)
sns.distplot(xi, bins = 17)
```

# Binomial

# Bonus distribution, F distribution

```python
xmin=.0
xmax=10
d1=1
d2=1

x=np.linspace(xmin,xmax,100)
yp=scipy.stats.f.pdf(x,d1,d2)
plt.plot(x,yp,'o')
```

```python
yc=scipy.stats.f.cdf(x,d1,d2)
plt.plot(x,yc,'o')
```

```python
r=np.random.uniform(.0,.804,10000)
r.sort()
interl=scipy.interpolate.interp1d(yc, x)
xi=interl(r)
plt.hist(xi,bins=30)
```

```python
sns.distplot(xi, bins = 30)
```
