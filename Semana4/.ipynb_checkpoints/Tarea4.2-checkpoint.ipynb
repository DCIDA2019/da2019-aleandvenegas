```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
import scipy
from mpl_toolkits.mplot3d import Axes3D
```

# Distribuciones uniforme y no uniforme

```python
r=np.random.uniform(0,1,2000)
t=np.random.uniform(0,2*np.pi,2000)
plt.plot(r,t,'o',markersize=2)
```

```python
x=r*np.cos(t)
y=r*np.sin(t)
plt.plot(r,t,'o',markersize=2)
plt.plot(x,y,'o',markersize=2)
```

```python
r=np.random.uniform(-1,1,2000)
t=np.random.uniform(-1,1,2000)
plt.plot(r,t,'o',markersize=2)
plt.plot(x,y,'o',markersize=2)
```

```python
x=np.random.uniform(-1,1,2000)
y=np.random.uniform(-1,1,2000)
x1=x[x**2+y**2<=1]
y1=y[x**2+y**2<=1]
r=np.sqrt(x1**2+y1**2)
t=np.arctan2(y1,x1)
x1=r*np.cos(t)
y1=r*np.sin(t)
plt.plot(x,y,'o',markersize=3)
plt.plot(x1,y1,'o',markersize=2)
```

```python
plt.figure(figsize=(15,15))
ax=plt.axes(projection='3d',azim=120,elev=30)
r = np.random.uniform(0,1,2000)
t=np.random.uniform(0,np.pi,2000)
p=np.random.uniform(0,2*np.pi,2000)
x=r*np.cos(p)*np.sin(t)
y=r*np.sin(p)*np.sin(t)
z=r*np.cos(t)
ax.plot3D(x,y,z,'o',markersize=2);
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
```

```python
plt.figure(figsize=(15,15))
ax=plt.axes(projection='3d',azim=120,elev=30)
x = np.random.uniform(-1,1,2000)
y = np.random.uniform(-1,1,2000)
z = np.random.uniform(-1,1,2000)
x1=x[x**2+y**2+z**2<=1]
y1=y[x**2+y**2+z**2<=1]
z1=z[x**2+y**2+z**2<=1]
r=np.sqrt(x1**2+y1**2+z1**2)
t=np.arctan2(np.sqrt(x1**2+y1**2),z1)
p=np.arctan2(y1,x1)
x1=r*np.cos(p)*np.sin(t)
y1=r*np.sin(p)*np.sin(t)
z1=r*np.cos(t)
ax.plot3D(x1,y1,z1,'o',markersize=2);
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
```
